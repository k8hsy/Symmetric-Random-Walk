```python
import numpy as np
import itertools
import matplotlib.pyplot as plt
```

```python
#Parameters
M = 1 #동전의 개수
t = 500 # 동전을 던지는 횟수
```

```python
random_walk = [-1, 1]
steps = np.random.choice(random_walk, size = (M,t)).T
origin = np.zeros((1,M))
rw_paths = np.concatenate([origin, steps]).cumsum(axis = 0)
```

```python
plt.plot(rw_paths)
plt.xlabel("Number of Flips")
plt.ylabel("Move")
plt.show()
```
